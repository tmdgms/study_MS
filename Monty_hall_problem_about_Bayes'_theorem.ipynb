# Monty Hall problem

http://j1w2k3.tistory.com/m/1009

#### 처음 이 문제가 공개 되었을 때 엄청난 이슈가 있었다. 일반적으로 이전의 나를 포함해 확률을 공부하지 않았던 사람들이라면 이 문제를 보고 이러한 생각을 할 것 이다.
- 처음 맞출 확률은 1/3 이었고 다른 한 문에 염소가 있다는 것을 알게 되면 확률은 1/2로 주니깐 내가 선택을 바꾸는 안바꾸든 1/2 이야!

#### 하지만 베이즈정리를 공부한뒤 정리해보면 확률은 1/3에서 1/3이 된다는 것을 알수 있을 것 이다. 왜 그러한지 차근차근 생각해보자.

##### <정리>
- 먼저 처음 정리하는 것이 중요하다. 우리는 한가지의 가정과 2가지의 event(사건)으로 나누어보고 우리가 알고 싶은것도 정리해보자.
    - 문은 A,B,C라고 명칭하자.
    - 가정 : 내가 선택한 문은 A 이다.
    - 사건1 : P(A),P(B),P(C)라는 말은 각각 A,B,C 문 뒤에 자동차가 있을 확률
    - 사건2 : 사건 D = "C문을 연다" 따라서 P(D) = C문을 열 확률
    - 우리가 알고 싶은 것 P(A) 에서 P(A|D) 이말은 A문뒤에 자동차가 있을 확률에서 D라는 사건 이후에 A문뒤에 자동차가 있을 확률

##### <재료 준비 (값)>
- 각각의 값을 생각해보면 P(A) = 1/3, P(B) = 1/3, P(C) = 1/3
- 베이즈정리를 생각해 봤을때 P(D|A)라는 말은 A 문뒤에 자동차가 있을때 사회자가 C문을 열 확률 이라는 말이다.
- 우리는 A를 선택하였고 A에 자동차가 있다고 가정하면 사회자가 열수있는 문은 B와 C 둘 중 하나이다.
    - 그러므로 P(D|A) = 1/2
- P(D|B)라는 말도 위와 같이 생각하자
- 우리는 A를 선택하였고 B에 자동차가 있다고 가정하면 사회자가 열수있는 문은 C하나 밖에 없다.
    - 그러므로 P(D|B) = 1
- P(D|C)라는 말도 위와 같이 생각하자.
- 우리는 A를 선택하였고 C에 자동차가 있다고 가정하면 사회자가 열수있는 문은 B하나 밖에 없다.
    - 그러므로 P(D|C) = 0

##### <공식을 재료에 맞춘후에 공식에 대입>
- 여기까지하면 재료준비가 다끝났다 이제 베이즈 정리 공식을 이용하여 값을 구하면 된다.
- 우리가 알고 싶은 것은 P(A|D)이다. 이 식을 재료에 맞게 변형시켜보자
    - P(A|D)
    - = P(D|A)*P(A)  /  P(D)
    - = P(D|A)*P(A)  /  P(D,A)+P(D,B)+P(D,C)
    - = P(D|A)*P(A)  /  P(D|A)P(A)+P(D|B)P(B)+P(D|C)P(C)
- 이 식에 맞춰서 재료를 넣어주면
    - (1/2 * 1/3) / (1/2 * 1/3) + (1 * 1/3) + (0 * 1/3)
    - = 1/3 (이 결과는 D사건 이후에 A라는 선택을 고수했을때의 확률이다.)
    
### <결론> : D사건이후 선택지는 두개니깐 선택을 바꾼다면 확률은 2/3로 증가

#### 확률을 공부함으로서 내 생각을 부순 케이스라서 이렇게 정리해 보았다.
